```python
import pandas as pd
import numpy as np
```

```python
url = "https://en.wikipedia.org/wiki/List_of_cities_in_India_by_population"
```

```python
data = pd.read_html(url)
```

```python
type(data)
```

```python
len(data)
```

```python
data[0]
```

```python
df = data[0]
```

```python
df
```

```python
df.tail()
```

```python
if df is not None:
    df = df.replace('-', '')
```

```python
df
```

```python
df.info()
```

```python
df['Rank'] = df['Rank'].str.replace('/', '')
```

```python
df['Rank'] = df['Rank'].astype(float)
```

```python
df['Population (2011)[3]'] = df['Population (2011)[3]'].str.replace(',', '').str.replace(r'\[.*\]', '')
```

```python
df['Population (2011)[3]'] = df['Population (2011)[3]'].astype(float)
```

```python
df['Population (2001)'] = pd.to_numeric(df['Population (2001)'], errors='coerce')
```

```python
 df.info()
```

```python
df.to_csv('List of cities in India by population.csv', index=False)
```
